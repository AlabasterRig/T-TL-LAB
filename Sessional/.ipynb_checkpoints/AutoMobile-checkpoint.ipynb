# Q3

```python
while True:
    num = input("Enter number more than 4 digits: ")
    if len(num) > 4:
        break
    
sum_digits = sum(int(digit) for digit in num)

while sum_digits > 9:
    sum_digits = sum(int(digit) for digit in str(sum_digits))
print("Sum of digits: ", sum_digits)
```

```python
plt.figure(figsize=(8, 6))  
plt.hist(Dataset['AnnualSalary'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.grid(True)  
plt.show()

# Creating a bar chart
plt.figure(figsize=(10, 6)) 
job_roles_counts = Dataset['JobTitle'].value_counts()
plt.bar(job_roles_counts.index, job_roles_counts.values, color='lightgreen')
plt.xlabel('Job Title')
plt.ylabel('Number of Employees')
plt.title('Number of Employees per Job Title')
plt.xticks(rotation=90) 
plt.grid(axis='y')  
plt.tight_layout()  
plt.show()
```
