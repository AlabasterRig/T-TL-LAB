# Pandas Series

A Pandas Series is like a column in a table.
It is a one-dimensional array holding data of any type.

```python
import pandas as pd

a = ["CS", "IT", "CSCE"]
branch = pd.Series(a)
print(branch)
```

# Lables

If noting else is specified, the values are labled with their index number. First value has index 0, second value has index 1, etc.

# Create Labels

```python
a = ["CS", "IT", "CSCE"]
branch = pd. Series(a, index=['a', 'b', 'c'])
print(branch)
print(branch['b'])
```

# Data Frames
A pandas dataframe is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns

```python
import pandas as pd
data = {
    'empid': [1,2,3],
    'salary': [50000, 40000, 45000]
}

df = pd.DataFrame(data)
print(df)
```

# Local Row
Pandas use the loc attribute to return one or more specified rows

```python
print(df.loc[0])
```

```python
print(df.loc[[0, 1]])
```
