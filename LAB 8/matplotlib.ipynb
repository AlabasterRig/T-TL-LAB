# Question 4
How to Fill Between Multiple Lines in Matplotlib?


```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='Sin')
plt.plot(x,y2,label='Cos')
plt.fill_between(x,y1,y2,color='grey',alpha=0.3)
plt.legend()
plt.show()
```

# Question 6
Display the value of each bar in a bar chart using Matplotlib?

```python
categories=['A','B','C','D','E']
values=[23,26,9,25,90]
bars=plt.bar(categories,values)
[plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),value,ha='center',va='bottom') for bar,value in zip(bars,values)]
plt.show()
```

# Question 13
How To Annotate Bars in Barplot with Matplotlib in Python?


```python
categories=['A','B','C','D','E']
values=[23,26,9,25,90]
plt.bar(categories,values)
for category,value in zip(categories,values):
    plt.text(category,value,str(value),ha='center',va='bottom')
plt.show()
```

# Question 14
How to Annotate Bars in Grouped BarplotinPython?

```python
categories=['A','B','C','D']
values1=[23,26,9,25]
values2=[6,3,4,9]
bar_width=0.35
x=np.arange(len(categories))
plt.bar(x-bar_width/2,values1,width=bar_width,label="Group 1")
plt.bar(x+bar_width/2,values2,width=bar_width,label="Group 2")
for i in range(len(categories)):
    plt.text(x[i]-bar_width/2,values1[i],str(values1[i]),ha='center',va='bottom')
    plt.text(x[i]+bar_width/2,values2[i],str(values2[i]),ha='center',va='bottom')
plt.show()
```

# Question 15

How to Display an Image in Grayscale in Matplotlib?


```python
import matplotlib.image as mpimg
image_path='cat.jpg'
image=mpimg.imread(image_path)
grayimg=image.mean(axis=2)
plt.imshow(grayimg,cmap='gray')
plt.axis('off')
plt.show()
```

# Question 17
How to Draw Rectangle on Image in Matplotlib?


```python
import matplotlib.patches as patches
image_path='cat.jpg'
image=mpimg.imread(image_path)
fig,ax=plt.subplots()
ax.imshow(image)
ax.axis('off')
rect=patches.Rectangle((50,50),500,900,linewidth=2,edgecolor='r',facecolor='none')
ax.add_patch(rect)
plt.show()
```

# Question 18
How to Display an OpenCV image in Python with Matplotlib?

```python
import cv2

image = cv2.imread('image.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
